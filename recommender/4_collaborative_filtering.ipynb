```python
# load data
file = open('dataset/processed_dataset/pivot_tab.pkl', 'rb')
pivot_tab = pickle.load(file)
```

```python
# pivot_tab = rating_df.pivot_table(index=['user_id'], columns=['anime_id'], values='rating').fillna(0)
pivot_tab.head()
```

```python
pivot_tab = pivot_tab.T
```

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim_pivot = csr_matrix(pivot_tab.values)
```

```python
item_similarity = cosine_similarity(cosine_sim_pivot)
user_similarity = cosine_similarity(cosine_sim_pivot.T)
item_sim_df = pd.DataFrame(item_similarity, index = pivot_tab.index, columns = pivot_tab.index)
user_sim_df = pd.DataFrame(user_similarity, index = pivot_tab.columns, columns = pivot_tab.columns)
```

```python
def get_rec_collaborative_anime_id(anime_id):
    rec = item_sim_df.sort_values(by = anime_id, ascending = False).index[1:11]
    return pd.DataFrame(rec)
```

```python
def get_rec_collaborative_user_id(user_id):
    rec = user_sim_df.sort_values(by = user_id, ascending = False).index[1:11]
    return pd.DataFrame(rec)
```

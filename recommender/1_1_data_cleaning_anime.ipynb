# ANIME RECOMMENDER SYSTEM - DATA CLEANING - ANIME

```python
# basic library
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import pickle
warnings.filterwarnings('ignore')
```

```python
# load data
anime_df = pd.read_csv('dataset/anime.csv')
anime_df.head()
```

## CLEANING NULL - DUPLICATES

```python
anime_df.info()
```

```python
anime_df['ranked'].replace({np.nan: 0}, inplace=True)
anime_df['title'].replace({np.nan: '-'}, inplace=True)
anime_df['synopsis'].replace({np.nan: 'Unknown'}, inplace=True)
```

```python
anime_df.info()
```

```python
anime_df = anime_df.drop_duplicates()
```

```python
anime_df.info()
```

```python
# drop because all unknown
anime_df = anime_df.drop(['season_aired'], axis=1)
```

## CLEANING DATA TYPES AND INVALID VALUES

```python
anime_df.rename(columns={'anime_rating': 'score'}, inplace=True)
print(sorted(pd.unique(anime_df['score'])))
```

```python
def to_num(col):
    # replace 'Unknown' to 0
    anime_df[col].replace({'Unknown': 0}, inplace=True)
    anime_df[col].replace({'?': 0}, inplace=True)

    # convert to float
    anime_df[col] = pd.to_numeric(anime_df[col], downcast='float')
```

```python
col_to_num_list = ['episode_count', 'score', 'rating_count']
for i in col_to_num_list:
    to_num(i)
```

```python
anime_df.info()
```

```python
anime_df.head()
```

```python
# list cleaning (string)
anime_df['genre'] = anime_df['genre'].str.replace('[','')
anime_df['genre'] = anime_df['genre'].str.replace(']','')
anime_df['genre'] = anime_df['genre'].str.replace('\'','')
```

```python
anime_df.head()
```

## EXPORT TO CSV

```python
anime_df.to_csv('dataset/cleaned_dataset/anime.csv', index=False)
```

```python
with open('./dataset/cleaned_dataset/anime.pkl','wb') as f:
    pickle.dump(anime_df, f)
```

```python
list_user_id = []
list_score = []
list_num_watched_episodes = []
list_created_at = []
list_updated_at = []
list_anime_id = []
```

```python
for i, row in anime_df.iterrows():
    list_user_id.append(1500022)
    list_score.append(0.742121)
    list_anime_id.append(row.anime_id)
```

```python
list_tuples = list(zip(list_user_id, list_score, list_anime_id)) 
```

```python
rating_dummy = pd.DataFrame(list_tuples, columns=['user_id', 'user_score', 'anime_id']) 
```

```python
rating_dummy.head()
```

```python
rating_df = rating_df.append(rating_dummy)
```

```python
rating_df.rename(columns = {'score':'user_score'}, inplace = True)
```

```python
merge_df = pd.merge(rating_df, anime_df, how='left')
merge_df.head()
```

```python
merge_df.count()
```

```python
pivot_tab = rating_df.pivot_table(index=['user_id'], columns=['anime_id'], values='user_score').fillna(0)
pivot_tab.head()
```

```python
merge_df.to_csv('dataset/processed_dataset/merged.csv', index=False)
```

```python
with open('./dataset/processed_dataset/pivot_tab.pkl','wb') as f:
    pickle.dump(pivot_tab, f)
```
